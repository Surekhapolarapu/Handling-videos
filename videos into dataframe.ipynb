```python
folders = []
for i in y :
    if i.endswith(".mp4") == False and i.endswith(".csv" )==False:
        folders.append(i)

folders
    
```

```python
features = []
class_labels = []
for folder in folders:
    for images in os.listdir(r"C:\Users\laksh\Downloads\cat_dog\{}".format(folder)):
        img = cv2.imread(r"C:\Users\laksh\Downloads\cat_dog\{}\{}".format(folder,images),0)
        img = cv2.resize(img,(50,50))
        img = img.flatten()
        features.append(img)
        class_labels.append(folder)
```

```python
data=pd.DataFrame(features )
data
```

```python
data['class_labels']=class_labels
data
```

```python
## save in csv file
data.to_csv(r"C:\Users\laksh\Downloads\cat_dog.csv")
```

### video capture with live stream

```python
vid = cv2.VideoCapture(0)

while True:
    succ,imag = vid.read()
    if succ == False:
        print("cam not working")
        break

    cv2.imshow("live stream",imag)

    if cv2.waitKey(1) & 255 == ord("A"):
        break

cv2.destroyAllWindows()
```

```python
# converting video in gray colour
vid = cv2.VideoCapture(0)

while True:
    succ,imag = vid.read()
    if succ == False:
        print("cam not working")
        break
    cv2.imshow("live stream",imag)    
    img1=cv2.cvtColor(imag,cv2.COLOR_BGR2GRAY)
    cv2.imshow("live stream",img1)
    

    if cv2.waitKey(1) & 255 == ord("A"):
        break

cv2.destroyAllWindows()
```

```python
# converting video in green,blue,red colour
vid = cv2.VideoCapture(0)

while True:
    succ,imag = vid.read()
    if succ == False:
        print("cam not working")
        break
    cv2.imshow("live stream",imag)    
    b,g,r = cv2.split(imag)
    z = np.zeros(b.shape, dtype = np.uint8)
    cv2.imshow("live stream",cv2.merge([b,z,z]))
    cv2.imshow("live stream",cv2.merge([z,g,z]))
    cv2.imshow("live stream",cv2.merge([z,z,r]))
    

    if cv2.waitKey(1) & 255 == ord("A"):
        break

cv2.destroyAllWindows()
```
